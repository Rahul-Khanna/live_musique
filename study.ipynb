```python
URI_dict_for_artist = {}
def othername_cleaner(input_text):
    input_list = re.split(',|\n',input_text)
    result = []
    for item in input_list:
        new_item = re.sub('\[1\]|\[2\]|\[3\]|\[4\]|\[5\]|"', '', item)
        new_item = new_item.strip()
        if '(' in new_item:
            index = new_item.find('(')
            new_item = new_item[:index].strip()
        result.append(new_item)
    return result
```

```python
def build_YearlyTouringInformation(g, wiki_url, sk_name, tour_hostory):
    # getting the Years active of artist -> str
    # create an instance of YearlyTouringInformation: URI is artist-year
    # get it from tour_hostory -> {'year(str)':times(int)}  and sk_name(str) is the name of artist 
    # sk_name need to be cleaning by replacing ' and " with '' and replace " " by ''
    # year need to change from str to gyear but no idea how to do it, so I change it to int
    for key in tour_hostory:
        sk_name = re.sub(r'[\'\"]*', '', sk_name)
        sk_name = sk_name.replace(' ', '-')
        uri = URIRef(my_ns[sk_name+'-'+key])
        g.add((wiki_url, my_ns.yearlyTouringData, uri)) # for entity class
        
        # create extra class for tour history
        g.add((uri, RDF.type, my_ns.YearlyTouringInformation))
        ####################################################################
        ########################## xsd: gYear ##############################
        ####################################################################
        g.add((uri, dbo.year, Literal(int(key))))
        
        g.add((uri, my_ns.concertCount, Literal(tour_hostory[key])))
    return g
```

```python
def build_playsOftenIn(g, wiki_url, often_play_in):
    # getting the Years active of artist -> str
    # create an instance of city
    # get it from often_play_in -> [['city(str)',times(int)],...] 
    # city need to clean by replacing ' ' with '-' and wipe out () by find the '(' and str[:indexof'(']
    # no need to clean when city is a value
    for pair in often_play_in:
        
        if '(' not in pair[0]:
            city = URIRef(my_ns[pair[0].replace(' ', '-')])
            city_lit = Literal(pair[0])
        else: 
            temp_city = pair[0].split('(')[0].strip()
            city = URIRef(my_ns[temp_city.replace(' ', '-')])
            city_lit = Literal(temp_city)
            
        g.add((wiki_url, my_ns.playsOftenIn, city)) # for the entity
        # create an instance for city
        g.add((city, RDF.type, schema.City))
        g.add((city, schema.name, city_lit))
    return g
```

```python
def build_Years_active(g, wiki_url, year_active_dict):
    # getting the Years active of artist -> str
    # get it from an_artist -> {'Years active':{text:'str',links:[]}} (might not exist)
    # the links is empty we don't need it ( the else is for insurance, make sure won't have an exception)
    # text is in format xxxx-present
    # need to split by '-' and only need the first element (year) in the splitting list
    if year_active_dict['links']:
        for item in year_active_dict['links']:
            year = item['text'].split('-')[0].strip()
            ##################### transfer to date ################
            cop = re.compile("[^0-9^-]")
            year = cop.sub('', year)[0:4]
            if year:
                year = datetime.strptime(year, '%Y').year
                g.add((wiki_url, schema.activeYearsStartYear, Literal(year))) # for the entity
        
    else:
        year = year_active_dict['text'].split('-')[0].strip()
        cop = re.compile("[^0-9^-]")
        year = cop.sub('', year)[0:4]
        year = datetime.strptime(year, '%Y').year
        if year:   
            g.add((wiki_url, schema.activeYearsStartYear, Literal(year))) # for the entity
    return g
```

```python
def build_genre(g, wiki_url, genre_dict):
    
    # getting the genre of artist -> str
    # also create an instance of genre
    # get it from genre_dict -> {text:'str',links:[{'text':'str',link:'str'}...]} 
    # check is there a link under the genre by if links
    # if there are info in the links then use links data 
    # link['link'] as URI 
    # need to clean the URI by replace ' ' by '-' 
    # link['text'] as value of name
    # if there is no info in the links then use text data 
    # need to clean the data by spliting it on '\' 
    # after splitting it for each item in the list would be the URI and the value of name
    # need to clean the URI by replace ' ' by '-'
    
    if genre_dict['links']:
        for item in genre_dict['links']:
            
            genre = URIRef(my_ns[[item['link'].replace(' ','-')]])
            
            g.add((wiki_url, dbo.genre, genre)) # for the entity class
            
            # create an instance for genre
            g.add((genre, RDF.type, dbo.MusicGenre))
            g.add((genre, schema.name, Literal(item['text'])))
            
    else:
        item = genre_dict['text']
        items = item.split("\\")
        for genre in items:
            
            genre_lit = Literal(genre.strip())
            genre = re.sub(r'\n','-',genre) # clean \n
            genre = URIRef(my_ns[genre.strip().replace(' ','-')])
            
            g.add((wiki_url, dbo.genre, genre)) # for the entity class
            
            # create an instance for label
            g.add((genre, RDF.type, dbo.MusicGenre))
            g.add((genre, schema.name, genre_lit))
    return g
```

```python
def build_record_label(g, wiki_url, label_dict):
    
    # THIS IS SAME TO Genres
    # getting the Labels of artist -> str
    # also create an instance of Labels
    # get it from label_dict -> {text:'str',links:[{'text':'str',link:'str'}...]}
    # check is there a link under the genre by if links
    # if there are info in the links then use links data 
    # link['link'] as URI 
    # need to clean the URI by replace ' ' by '-' 
    # link['text'] as value of name
    # if there is no info in the links then use text data 
    # need to clean the data by spliting it on '\' 
    # after splitting it for each item in the list would be the URI and the value of name
    # need to clean the URI by replace ' ' by '-'
    
    if label_dict['links']:
        for item in label_dict['links']:
            
            if 'Rimas' in item['link']:
                print(item['link'])
                
            label = URIRef(my_ns[item['link'].replace(' ','-')])
            
            g.add((wiki_url, dbo.recordLabel, label)) # for the entity class
            
            # create an class for label
            g.add((label, RDF.type, dbo.RecordLabel))
            g.add((label, schema.name, Literal(item['text'])))
    else:
        item = label_dict['text']
        items = item.split("\\")
        for label in items:
            
            label_lit = Literal(label)
            label = re.sub(r'\n','-',label) # clean \n
            label = URIRef(my_ns[label.replace(' ','-')])
            
            g.add((wiki_url, dbo.recordLabel, label)) # for the entity class
            
            # create extra class for label
            g.add((label, RDF.type, dbo.RecordLabel))
            g.add((label, schema.name, label_lit))
    return g
```
