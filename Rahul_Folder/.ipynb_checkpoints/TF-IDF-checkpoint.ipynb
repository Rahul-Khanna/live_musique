```python
nlp = spacy.load("en_core_web_md")
```

```python
urls = list(artist_reviews.keys())
```

```python
text_data = []
for url in tqdm.tqdm(urls):
    doc = artist_reviews[url]["text"]
    kept_tokens = []
    for token in doc:
        if token.pos_ == "ADJ" and not token.is_stop:
            kept_tokens.append(token.text.lower())
    text = " ".join(kept_tokens)
    text_data.append(text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
```

```python
tokenizer = lambda x: x.split()
```

```python
vectorizer = TfidfVectorizer(lowercase=True, tokenizer=tokenizer, ngram_range=(1,1), max_df=0.4, sublinear_tf=True)
```

```python
X = vectorizer.fit_transform(text_data)
```

```python
X = X.tocoo()
```

```python
def sort_coo(m):
    tuples = zip(m.row, m.col, m.data)
    return sorted(tuples, key=lambda x: (x[0], x[2]), reverse=True)
```

```python
data_X = sort_coo(X)
```

```python
data_X[0]
```

```python
top_words_per_url = {}
```

```python
features = vectorizer.get_feature_names()
```

```python
"and" in features
```

```python
for i, tup in enumerate(tqdm.tqdm(data_X)):
    url = urls[tup[0]]
    phrase = features[tup[1]]
    score = tup[2]
    if url in top_words_per_url:
        if len(top_words_per_url[url]) < 10:
            top_words_per_url[url].append((phrase, score))
    else:
        top_words_per_url[url] = [(phrase, score)]
```

```python
with open("nlp_data/top_adj_sg_url.p", "wb") as f:
    pickle.dump(top_words_per_url, f)
```
